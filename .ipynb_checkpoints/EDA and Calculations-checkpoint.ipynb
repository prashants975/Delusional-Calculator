```python
import pandas as pd

data = {
    'Range (in INR)': ['< 0', '= 0', '>0 and <=1,50,000', '>1,50,000 and <= 2,00,000', '>2,00,000 and <= 2,50,000',
                       '>2,50,000 and <= 3,50,000', '>3,50,000 and <= 4,00,000', '>4,00,000 and <= 4,50,000',
                       '>4,50,000 and <= 5,00,000', '>5,00,000 and <= 5,50,000', '>5,50,000 and <= 9,50,000',
                       '>9,50,000 and <= 10,00,000', '>10,00,000 and <=15,00,000', '>15,00,000 and <= 20,00,000',
                       '>20,00,000 and <= 25,00,000', '>25,00,000 and <= 50,00,000', '>50,00,000 and <= 1,00,00,000',
                       '>1,00,00,000 and <=5,00,00,000', '>5,00,00,000 and <=10,00,00,000',
                       '>10,00,00,000 and <=25,00,00,000', '>25,00,00,000 and <=50,00,00,000',
                       '>50,00,00,000 and <=100,00,00,000', '>100,00,00,000 and <=500,00,00,000',
                       '>500,00,00,000'],
    'No. of Returns': ['-', '1,70,767', '23,78,193', '13,76,970', '37,87,092', '1,42,75,685', '50,25,936',
                       '41,56,142', '37,83,374', '29,99,708', '1,11,53,567', '6,00,970', '30,01,781', '9,98,299',
                       '5,08,642', '7,36,059', '2,09,345', '89,793', '5,132', '2,089', '451', '147', '74', '3'],
    'Sum of Gross Total Income (In Crore INR)': ['-', '-', '0.82', '1.77', '2.34', '3.05', '3.73', '4.25', '4.75', '5.23',
                                                 '7.06', '9.75', '11.96', '17.17', '22.24', '33.66', '67.69', '179.46',
                                                 '678.38', '1,486.15', '3,431.07', '6,754.40', '17,743.08', '78,134.03'],
    'Average Gross Total Income (In Lakh INR)': ['-', '-', '0.82', '1.77', '2.34', '3.05', '3.73', '4.25', '4.75', '5.23',
                                                 '7.06', '9.75', '11.96', '17.17', '22.24', '33.66', '67.69', '179.46',
                                                 '678.38', '1,486.15', '3,431.07', '6,754.40', '17,743.08', '78,134.03']
}

df_income = pd.DataFrame(data)

# Function to convert numbers with commas to integers
def convert_to_int(x):
    return int(x.replace(',', '')) if x != '-' else None

# Function to convert numbers with commas to floats
def convert_to_float(x):
    return float(x.replace(',', '')) if x != '-' else None

# Apply conversion to relevant columns
df_income['No. of Returns'] = df_income['No. of Returns'].apply(convert_to_int)
df_income['Sum of Gross Total Income (In Crore INR)'] = df_income['Sum of Gross Total Income (In Crore INR)'].apply(convert_to_float)
df_income['Average Gross Total Income (In Lakh INR)'] = df_income['Average Gross Total Income (In Lakh INR)'].apply(convert_to_float)

```

```python
df_income.drop(0, inplace = True)
df_income.drop(23, inplace = True)
df_income.reset_index(drop = True)
```

```python
df_income.fillna(0, inplace = True)
df_income["Percenatge in Income Range"] = df_income['No. of Returns']/df_income['No. of Returns'].sum()*100
```

```python
import re
def remove_special_characters(text):
    # Regular expression to match any character that is not alphanumeric or a space
    pattern = r'[^a-zA-Z0-9\s]'
    # Replace the special characters with an empty string
    return re.sub(pattern, '', text)

# Apply the function to the 'special_column' in the DataFrame
df_income["Income Up Boundary"] = df_income["Range (in INR)"].str.split("and").str[-1].apply(remove_special_characters).astype(float)
```

```python
df_income['CDF'] = df_income['Percenatge in Income Range'].cumsum()
df_income['Cum_Returns'] = df_income['No. of Returns'].cumsum()
```

```python
from scipy.stats import norm

height_stats = {
    "Male": [170, 6.35],
    "Female": [158, 5.59]
}


def calculate_cdf(x, mean, std_dev):
    z_score = (x - mean) / std_dev
    cdf = norm.cdf(z_score)
    return cdf

def height_percentage(gender, x_value_min, x_value_max):

    mean = height_stats[gender][0]
    std_dev = height_stats[gender][1]
    
    
    h_percent = calculate_cdf(x_value_max, mean, std_dev) - calculate_cdf(x_value_min, mean, std_dev)
    return h_percent
```

```python
height_percentage("Male", 182, 190)
```
